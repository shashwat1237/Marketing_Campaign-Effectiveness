```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("marketing_campaign_dataset.csv",header=0)
df.info()
df.isnull().sum()
```

```python
df["Duration"].unique()
df.head()
```

```python
def convert_duration(duration):

    return int((duration).replace(" days",""))

def convert_acquisition(acquisition_cost):

    return float(acquisition_cost.replace("$","").replace(",",""))


df["Duration"]=convert_duration(df["Duration"])
df["Acquisition_Cost"]=convert_acquisition(df["Acquisition_Cost"])
    
```
